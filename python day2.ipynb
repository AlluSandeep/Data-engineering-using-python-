```python
std1={'sid':1011,'name':"abhinay",'branch':"cse"}
std2={'sid':1012,'name':"sandeep",'branch':"ece"}
std3={'sid':1013,'name':"siva ram",'branch':"mech"}
std4={'sid':1014,'name':"lohit",'branch':"civil"}
std5={'sid':1015,'name':"pallavi",'branch':"cse"}
std6={'sid':1016,'name':"monika",'branch':"cse"}
std7={'sid':1017,'name':"greeshma",'branch':"ece"}
std8={'sid':1018,'name':"pranavi",'branch':"mech"}
std9={'sid':1019,'name':"sravani",'branch':"civil"}
std10={'sid':1020,'name':"sony",'branch':"cse"}
stdlist=[std1,std2,std3,std4,std5,std6,std7,std8,std9,std10]
ac=0
mc=0
cc=0
cvc=0
for std in stdlist:
  if std.get('branch')=='cse':
    ac=ac+1
  elif std.get('branch')=='ece':
    cc=cc+1
  elif std.get('branch')=='mech':
    mc=mc+1
  elif std.get('branch')=='civil':
    cvc=cvc+1
print(ac,cc,mc,cvc)
```

```python
import csv
fp=open('/content/drive/My Drive/sandeep/empdata1.csv','r')
data=csv.DictReader(fp)
hc=[]
fc=[]
mc=[]
for row in data:
  if row.get('dname')=='hr':
    hc.append(eval(row.get('sal')))
  elif row.get('dname')=='fin':
    fc.append(eval(row.get('sal')))
  elif row.get('dname')=='mark':
    mc.append(eval(row.get('sal')))
print(sum(hc),max(hc),min(hc),sum(hc)/len(hc))
print(hc,fc,mc)
fp.close()
```
