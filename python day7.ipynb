```python
import re
string='welcome to python'
pattern="[^a-c]"
result=re.findall(pattern,string)
print(result)
```

```python
import re
string='welcome to python programming'
pattern="[a-c]"
result=re.findall(pattern,string)
print(result)
```

```python
import re
regex=r'programming$'
strings=['python programming','c programming']
for string in strings:
  if re.search(regex,string):
    print(f'Matched:{string}')
  else:
    print(f'Not Matched:{string}')
```

```python
import re
string="a#c"
pattern=r"a.c"
match=re.search(pattern,string)
if match:
  print("Match found")
else:
  print("Match not found")
```

```python
import re
string="zvvbc"
pattern=r"a/b"
match=re.search(pattern,string)
if match:
  print("Match found")
else:
  print("Match not found")
```
