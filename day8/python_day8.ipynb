```python
from math import sin
x=[0,30,45,60,90]
y=[sin(i) for i in x]
plt.plot(x,y)
plt.show()
```

```python
from math import sin
x=[0,30,45,60,90]
y=[sin(i) for i in x]
plt.xlabel("unit values")
plt.ylabel("sin values")
plt.title("sin graph")
plt.plot(x,y)
plt.show()
```

```python
from math import sin,cos
x=[0,30,45,60,90]
siny=[sin(i) for i in x]
cosy=[cos(i) for i in x]
plt.xlabel("unit values")
plt.ylabel("sin values")
plt.title("sin graph")
plt.plot(x,siny)
plt.plot(x,cosy)
plt.show()
```

```python
from math import sin,cos
x=[0,30,45,60,90]
siny=[sin(i) for i in x]
cosy=[cos(i) for i in x]
plt.xlabel("unit values")
plt.ylabel("sin values")
plt.title("sin graph")
plt.plot(x,siny,label="sin")
plt.plot(x,cosy,label="cos")
plt.legend()
plt.show()
```

```python
from math import sin,cos
x=[0,30,45,60,90]
siny=[sin(i) for i in x]
cosy=[cos(i) for i in x]
plt.xlabel("unit values")
plt.ylabel("sin values")
plt.title("sin graph")
plt.plot(x,siny,label="sin")
plt.plot(x,cosy,label="cos")
plt.legend(loc='right')
plt.show()
```
